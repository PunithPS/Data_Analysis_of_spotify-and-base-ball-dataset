```python
# importing the neccessary files for the analysis and reading the csv file and checking the data types of the each column
import numpy as np
import pandas as pd


df = pd.read_csv("Most Streamed Spotify Songs 2024_950803.csv",encoding='latin-1')

df.info()
```

```python
# analysing the data present in the dataset
df
```

```python
# converting the data types of the neccessary column from object type to float , which helps in analysis in future
df['Spotify Streams'] = df['Spotify Streams'].str.replace(',','').astype(float)
df['Spotify Playlist Count'] = df['Spotify Playlist Count'].str.replace(',','').astype(float)
df['Spotify Playlist Reach'] = df['Spotify Playlist Reach'].str.replace(',','').astype(float)
df['YouTube Views'] = df['YouTube Views'].str.replace(',','').astype(float)
df['YouTube Likes'] = df['YouTube Likes'].str.replace(',','').astype(float)
df['TikTok Posts'] = df['TikTok Posts'].str.replace(',','').astype(float)
df['TikTok Likes'] = df['TikTok Likes'].str.replace(',','').astype(float)
```

```python
# converting the data types of the neccessary column from object type to float , which helps in analysis in future
df['TikTok Views'] = df['TikTok Views'].str.replace(',','').astype(float)
df['YouTube Playlist Reach'] = df['YouTube Playlist Reach'].str.replace(',','').astype(float)
df['AirPlay Spins'] = df['AirPlay Spins'].str.replace(',','').astype(float)
df['SiriusXM Spins'] = df['SiriusXM Spins'].str.replace(',','').astype(float)
df['Deezer Playlist Reach'] = df['Deezer Playlist Reach'].str.replace(',','').astype(float)
df['Pandora Streams'] = df['Pandora Streams'].str.replace(',','').astype(float)
df['Pandora Track Stations'] = df['Pandora Track Stations'].str.replace(',','').astype(float)
df['Soundcloud Streams'] = df['Soundcloud Streams'].str.replace(',','').astype(float)
df['Shazam Counts'] = df['Shazam Counts'].str.replace(',','').astype(float)
```

```python
df.info()
```

```python
#Deleating the duplicate entries from the data aset
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
# checking the presence of null value
df.isnull().sum()
```

```python
# Handling the missing values of the column by using the forward fill for the artist column and
#adding the avg or the mean values to missing values of the that perticular column
df['Artist'].fillna(method='ffill',inplace=True)
df['Spotify Streams'].fillna(df['Spotify Streams'].mean())
df['Spotify Playlist Count'].fillna(df['Spotify Playlist Count'].mean())
df['Spotify Playlist Reach'].fillna(df['Spotify Playlist Reach'].mean())
df['Spotify Popularity'].fillna(df['Spotify Popularity'].mean())
df['YouTube Views'].fillna(df['YouTube Views'].mean())
df['YouTube Likes'].fillna(df['YouTube Likes'].mean())
df['TikTok Posts'].fillna(df['TikTok Posts'].mean())
df['TikTok Likes'].fillna(df['TikTok Likes'].mean())
df['TikTok Views'].fillna(df['TikTok Views'].mean())
df['YouTube Playlist Reach'].fillna(df['YouTube Playlist Reach'].mean())
df['Apple Music Playlist Count'].fillna(df['Apple Music Playlist Count'].mean())
df['AirPlay Spins'].fillna(df['AirPlay Spins'].mean())
df['SiriusXM Spins'].fillna(df['SiriusXM Spins'].mean())
df['Deezer Playlist Count'].fillna(df['Deezer Playlist Count'].mean())
df['Deezer Playlist Reach'].fillna(df['Deezer Playlist Reach'].mean())
df['Amazon Playlist Count'].fillna(df['Amazon Playlist Count'].mean())
df['Pandora Track Stations'].fillna(df['Pandora Track Stations'].mean())
df['Soundcloud Streams'].fillna(df['Soundcloud Streams'].mean())
df['Shazam Counts'].fillna(df['Shazam Counts'].mean())
```

```python
#average spotify streams count for all the track
avg_spotify_stream = df['Spotify Streams'].mean()
print(avg_spotify_stream)

avg_Sptfy_Plist_Count = df['Spotify Playlist Count'].mean()
print(avg_Sptfy_Plist_Count)
```

```python
#median Shazam Counts
Shazam_Counts_median = df['Shazam Counts'].median()
print("the median Shazam Counts is", Shazam_Counts_median)
```

```python
#Mode of the all time rank
all_time_mode = df['All Time Rank'].mode()
print(  "Mode of the all time rank is",all_time_mode)
```

```python
#Track that have explicit  track value 1
explicit_track = df[df['Explicit Track']==1]
print("Track that have explicit  track value 1 is",explicit_track)
```

```python
#converting the release date col to date time type from object type
df['Release Date'] = pd.to_datetime(df['Release Date'], errors = 'coerce')
```

```python
#can exctrct the year from whole input
df["Release Year"] = df['Release Date'].dt.year
```

```python
#avg spotify stream counts for track reased in year 2024
df_2024 = df[df['Release Date'].dt.year == 2024]
avg_spotify_stream_cnt = df_2024['Spotify Streams'].mean()
print("avg spotify stream counts for track reased in year 2024 is",avg_spotify_stream_cnt)
```

```python
#spotify stream counts for each artist
artist_filter = df.groupby('Artist')
total_spotify_stream_by_artist = artist_filter['Spotify Streams'].sum()
print("spotify stream counts for each artist is",total_spotify_stream_by_artist)
```

```python
#artist with highest avg track score
avg_score = df.groupby('Artist')['Track Score'].mean()
highest_avg_artist = avg_score.idxmax()

highest_avg_score = avg_score.max()
print(f'the artist with highest avg track score is{highest_avg_artist} with an avg score of{highest_avg_score}.')
```

```python
#the artist with highest Deezer Playlist Reach
artist_filter = df.groupby('Artist')['Deezer Playlist Reach'].sum()
top_artist = artist_filter.idxmax()
top_Deezer_reach = artist_filter.max()
print(f'the artist with highest Deezer Playlist Reach is{top_artist} with a reach of{top_Deezer_reach}.')
```

```python
#bar chart of top 10 trcaks by  Spotify Streams
import matplotlib.pyplot as plt

top_10_tracks = df.nlargest(10,'Spotify Streams')

plt.figure(figsize=(8,6))
plt.barh(top_10_tracks['Track'],top_10_tracks['Spotify Streams'])
plt.title("Top 10 Tracks on  Spotify")
plt.xlabel("Spotify Streams")
#plt.ylable("Track Name")
plt.show()
```

```python
# top 5 tracks by Shazam Counts
top_5_tracks = df.nlargest(5,'Shazam Counts')

plt.figure(figsize = (8,6))
plt.barh(top_5_tracks['Track'],top_5_tracks['Shazam Counts'])

plt.xlabel("Shazam Counts")
plt.ylabel('Track Name')
plt.title("Top 5 Tracks by Shazam")
plt.show()
```
